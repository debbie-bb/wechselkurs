```python
import requests

url = "http://127.0.0.1:8000/wechseln"  # Lokale URL deines Wechselkurs-API-Servers
payload = {"betrag": 100, "wechselkurs": 1.08}  # Beispielpayload
resp = requests.post(url, json=payload)
data = resp.json()
print(f"Ergebnis: {data['ergebnis']}")
```
